```python

df['date_of_filing'] = pd.to_datetime(df['date_of_filing'], format='%Y-%m-%d',errors = 'coerce') 


df['date_of_decision'] = pd.to_datetime(df['date_of_decision'], format='%Y-%m-%d',errors = 'coerce') 

df['date_next_list'] = pd.to_datetime(df['date_next_list'], format='%Y-%m-%d',errors = 'coerce') 

df['date_first_list'] = pd.to_datetime(df['date_first_list'], format='%Y-%m-%d',errors = 'coerce') 


```

```python

# df['file_end'] =(df['date_next_list']-df['date_of_filing']).dt.year
df.info()

df['closing'] = df['date_next_list'].dt.year
df['filing'] = df['date_of_filing'].dt.year
df['first'] = df['date_first_list'].dt.year
df['decision']=df['date_of_decision'].dt.year




```

```python
df['closing']=df['closing'].fillna('1900')
df['first']=df['first'].fillna('1900')
df['decision']=df['decision'].fillna('1900')
```

```python

df['closing']=df['closing'].astype(int)
df['filing']=df['filing'].astype(int)
df['first']=df['first'].astype(int)
df['decision']=df['decision'].astype(int)
```

```python

df1 = df[df['closing'].isin([2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022])]
```

```python
df2 = df1[df1['first'].isin([2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022])]
```

```python
df3=df2[df2['decision'].isin([2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022])]

df3
```

```python

df3['file_closing_days']=(df3['date_next_list']-df3['date_of_filing']).dt.days
df3['file_reach_days']=(df3['date_first_list']-df3['date_of_filing']).dt.days
df3['decide_close_days']=(df3['date_next_list']-df3['date_of_decision']).dt.days
```

```python

df3['file_closing_days']=df3['file_closing_days'].clip(lower=0)
df3['file_reach_days']=df3['file_reach_days'].clip(lower=0)
df3['decide_close_days']=df3['decide_close_days'].clip(lower=0)

#df['filing'].unique()

#df=df.drop(columns=['filing','closing'])
```

```python

df3=df3.reset_index()
df3=df3.drop(columns=['index'])
#=df.drop(columns=['index'])
df3
```

```python
df4=df3.drop(columns=['date_of_filing','date_of_decision','date_first_list','date_next_list','closing','filing','first','decision'])
```
