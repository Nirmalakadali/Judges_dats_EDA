```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.animation import FuncAnimation
%matplotlib notebook
%matplotlib inline
# Reading data
data = pd.read_csv('/home/nirmala/merged_final.csv')
data
# List of states
states = data['state'].unique()

# List of years
years = data['year'].unique()
years
# Data for each year
prison = data.groupby(['year'])['prison_count'].sum()
dismissed = data.groupby(['year'])['dismissed_count'].sum()
criminal = data.groupby(['year'])['criminal_count'].sum()
non_criminal = data.groupby(['year'])['noncriminal_count'].sum()
withdrawal=data.groupby(['year'])['withdrawal_count'].sum()
total_cases=data.groupby(['year'])['total_cases'].sum()
# Creating figure
fig, ax = plt.subplots(figsize=(10, 5))

# # Stacked bar chart
# def update(year):
#     ax.clear()
#     ax.bar(states, prison[year], label='Prison')
#     ax.bar(states, dismissed[year], bottom=prison[year], label='Dismissed')
#     ax.bar(states, criminal[year], bottom=prison[year]+dismissed[year], label='Criminal')
#     ax.bar(states, non_criminal[year], bottom=prison[year]+dismissed[year]+criminal[year], label='Non_Criminal')
#     ax.bar(states, withdrawal[year], bottom=prison[year]+dismissed[year]+criminal[year]+non_criminal[year], label='withdrawal')
#     ax.bar(states, total_cases[year], bottom=prison[year]+dismissed[year]+criminal[year]+non_criminal[year]+withdrawal[year], label='total_cases')
#     plt.xticks(states, rotation=90)
#     plt.yticks(np.arange(0, 80000, 10000))
#     plt.legend()
#     plt.title("Cases for year - {}".format(year))

# ani = FuncAnimation(fig, update, frames=years, repeat=True)
# plt.show()
def update(year):
    ax.clear()
    ax.bar(states, prison[year], label='Prison')
    ax.bar(states, dismissed[year], bottom=prison[year], label='Dismissed')
    ax.bar(states, criminal[year], bottom=prison[year]+dismissed[year], label='Criminal')
    ax.bar(states, non_criminal[year], bottom=prison[year]+dismissed[year]+criminal[year], label='Non_Criminal')
    ax.bar(states, withdrawal[year], bottom=prison[year]+dismissed[year]+criminal[year]+non_criminal[year], label='withdrawal')
    ax.bar(states, total_cases[year], bottom=prison[year]+dismissed[year]+criminal[year]+non_criminal[year]+withdrawal[year], label='total_cases')
    plt.xticks(states, rotation=90)
    plt.yticks(np.arange(0, 80000, 10000))
    plt.legend()
    plt.title("Cases for year - {}".format(year))

ani = FuncAnimation(fig, update, frames=years, repeat=True)
plt.show()
```

```python
for yer in list_of_years:
    df_gif = df[df.year==yer] 
    # Selecting only landslide category
# df=df.drop(columns=['prison_count','withdrawal_count','avg_decision_end_per1lakhcases','avg_fileend_per1lakhcases','avg_filecourt_per1lakhcases','pendingcases_count','criminal_count'])
    count='withdrawal_count'
    columns_to_keep = ['year', 'state',count]
    columns_to_drop = set(df.columns) - set(columns_to_keep)
    df_gif = df_gif.drop(columns=columns_to_drop, axis=1)
    df_dis=df_gif
    df_gif=df_gif.reset_index()
    df_merged=pd.merge(map_df,df_gif,on="state")
    df_merged=df_merged.drop(columns=['id','index'])
    fig, ax = plt.subplots(1, figsize=(10, 10))
    df_merged.plot(column='withdrawal_count',cmap='YlOrRd', linewidth=0.8, ax=ax, edgecolor='0', \
                legend=True,markersize=[39.739192, -104.990337], legend_kwds={'label': "Number of withdrawal_cases"})
    ax.axis('off')
    ax.set_title('Number of withdrawn cases in '+str(yer),fontdict={'fontsize': '20', 'fontweight' : '10'})
    filepath = os.path.join(output_path +'withdrawn '+ str(yer) + 'graph.jpg')
    
    chart = fig.get_figure()
    chart.savefig(filepath, dpi=300)
    images.append(imageio.imread(filepath.format(i)))
    i=i+1
    del df_merged
    del df_gif
```
